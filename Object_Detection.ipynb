```python
import os
import xml.etree.ElementTree as ET
import torch
from torch.utils.data import Dataset
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches

class CustomDataset(Dataset):
    def __init__(self, image_dir, annotation_dir, transform=None):
        self.image_dir = image_dir
        self.annotation_dir = annotation_dir
        self.image_paths = [
            os.path.join(image_dir, img) for img in os.listdir(image_dir) if img.endswith('.jpg') or img.endswith('.jpeg')
        ]
        self.annotation_paths = [
            os.path.join(annotation_dir, ann) for ann in os.listdir(annotation_dir) if ann.endswith('.xml')
        ]

        # Ensure that both lists have the same number of items
        assert len(self.image_paths) == len(self.annotation_paths), "Mismatch between images and annotations."

        self.transform = transform

        # Create a mapping of class names to indices
        self.class_map = {'car': 1, 'background': 0}  # Add more classes if needed

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Load image
        image_path = self.image_paths[idx]
        print(f'Loading image: {image_path}')  # Debug print
        image = Image.open(image_path).convert("RGB")

        # Load annotation
        annotation_path = self.annotation_paths[idx]
        print(f'Loading annotation: {annotation_path}')  # Debug print
        annotation = self.load_annotation(annotation_path)

        if self.transform:
            image = self.transform(image)

        return image, annotation

    def load_annotation(self, annotation_path):
        tree = ET.parse(annotation_path)
        root = tree.getroot()

        boxes = []
        labels = []
        for obj in root.findall('object'):
            label = obj.find('name').text
            bbox = obj.find('bndbox')
            x_min = int(bbox.find('xmin').text)
            y_min = int(bbox.find('ymin').text)
            x_max = int(bbox.find('xmax').text)
            y_max = int(bbox.find('ymax').text)
            boxes.append([x_min, y_min, x_max, y_max])
            # Convert label to numerical index
            if label in self.class_map:
                labels.append(self.class_map[label])
            else:
                raise ValueError(f"Label '{label}' not found in class map.")

        # Convert boxes and labels to tensors
        if boxes and labels:  # Ensure boxes and labels are not empty
            boxes = torch.tensor(boxes, dtype=torch.float32)
            labels = torch.tensor(labels, dtype=torch.long)  # Convert labels to tensor
            return {'boxes': boxes, 'labels': labels}
        else:
            # Return a default annotation if no boxes/labels are found
            return {'boxes': torch.empty((0, 4)), 'labels': torch.empty((0,), dtype=torch.long)}

def visualize_image_with_annotations(dataset, idx):
    image, annotations = dataset[idx]

    # Check if image is a tensor and convert accordingly
    if isinstance(image, torch.Tensor):
        image = Image.fromarray(image.permute(1, 2, 0).numpy().astype('uint8'))
    else:
        image = image.convert("RGB")  # Ensure image is in RGB format if it's already a PIL Image

    # Create a plot
    fig, ax = plt.subplots(1)
    ax.imshow(image)

    # Draw bounding boxes
    boxes = annotations['boxes'].numpy()
    labels = annotations['labels'].numpy()

    # Create a reverse mapping from index to class name
    reverse_class_map = {v: k for k, v in dataset.class_map.items()}

    for box, label in zip(boxes, labels):
        # Create a Rectangle patch
        rect = patches.Rectangle((box[0], box[1]), box[2] - box[0], box[3] - box[1],
                                 linewidth=1, edgecolor='r', facecolor='none')
        # Add the patch to the Axes
        ax.add_patch(rect)
        # Retrieve the class name from the class map
        class_name = reverse_class_map.get(label.item(), "Unknown")
        # Add the label
        ax.text(box[0], box[1], class_name, color='white', fontsize=12, bbox=dict(facecolor='red', alpha=0.5))

    plt.axis('off')  # Turn off axis numbers and ticks
    plt.show()


# Initialize the dataset and check
image_dir = '/content/dataset/valid'  # Change to your actual path
annotation_dir = '/content/dataset/valid'  # Change to your actual path

# Create the dataset
dataset = CustomDataset(image_dir=image_dir, annotation_dir=annotation_dir)

# Example: Check the length of the dataset
print(f'Dataset size: {len(dataset)}')

# Visualize the first image with annotations
visualize_image_with_annotations(dataset, 0)
```

```python
import os
import xml.etree.ElementTree as ET
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import files

class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, image_dir, annotation_dir, transform=None):
        self.image_dir = image_dir
        self.annotation_dir = annotation_dir
        self.image_paths = [os.path.join(image_dir, f) for f in os.listdir(image_dir) if f.endswith('.jpg')]
        self.annotation_paths = [os.path.join(annotation_dir, f) for f in os.listdir(annotation_dir) if f.endswith('.xml')]
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        annotation_path = self.annotation_paths[idx]

        # Load image
        image = Image.open(image_path).convert("RGB")

        # Load annotations
        annotations = self.load_annotation(annotation_path)

        if self.transform:
            image = self.transform(image)

        return image, annotations

    def load_annotation(self, annotation_path):
        tree = ET.parse(annotation_path)
        root = tree.getroot()

        boxes = []
        labels = []

        for obj in root.findall('object'):
            label = obj.find('name').text
            xmin = int(obj.find('bndbox/xmin').text)
            ymin = int(obj.find('bndbox/ymin').text)
            xmax = int(obj.find('bndbox/xmax').text)
            ymax = int(obj.find('bndbox/ymax').text)

            boxes.append([xmin, ymin, xmax, ymax])
            labels.append(label)  # Store the label as a string

        if boxes and labels:
            boxes = torch.tensor(boxes, dtype=torch.float32)
            return {'boxes': boxes, 'labels': labels}
        else:
            return {'boxes': torch.empty((0, 4)), 'labels': []}

# Function to visualize images with annotations
def visualize_image_with_annotations(dataset, idx):
    image, annotations = dataset[idx]

    # Convert the image back to PIL format for display
    image = Image.fromarray(image.permute(1, 2, 0).numpy().astype('uint8'))

    # Create a plot
    plt.figure(figsize=(8, 6))
    plt.imshow(image)

    # Draw boxes and labels
    for box, label in zip(annotations['boxes'], annotations['labels']):
        xmin, ymin, xmax, ymax = box.tolist()
        plt.gca().add_patch(plt.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin, fill=None, edgecolor='red', linewidth=2))
        plt.text(xmin, ymin, label, color='red', fontsize=12)

    plt.axis('off')
    plt.show()

# Define your image and annotation directories
image_dir = '/content/dataset/new_valid'  # Change to your actual path for new images
annotation_dir = '/content/dataset/new_valid'  # Change to your actual path for new annotations

# Create directories if they do not exist
os.makedirs(image_dir, exist_ok=True)
os.makedirs(annotation_dir, exist_ok=True)

# Check the contents of the dataset directory
print("Contents of the dataset directory:")
print(os.listdir(image_dir))

# Upload images and XML files
print("Upload your images and XML annotations:")
uploaded = files.upload()

# Create the dataset
transform = transforms.Compose([
    transforms.Resize((300, 300)),  # Resize images if needed
    transforms.ToTensor(),
])
dataset = CustomDataset(image_dir=image_dir, annotation_dir=annotation_dir, transform=transform)

# Example: Check the length of the dataset
print(f'Dataset size: {len(dataset)}')

# Visualize a random image with annotations if the dataset is not empty
if len(dataset) > 0:
    import random
    random_index = random.randint(0, len(dataset) - 1)
    visualize_image_with_annotations(dataset, random_index)
else:
    print("No images found in the dataset.")
```
